```python
# Student CGPA Calculator

# Student Details
name = input("Enter Student Name: ")
roll = input("Enter Roll Number: ")

# Subjects
subjects = ["Math", "Physics", "Chemistry", "English", "Computer Science"]

# Dictionary to store marks
marks = {}

# Taking marks input
for subject in subjects:
    mark = float(input(f"Enter marks in {subject} (out of 100): "))
    marks[subject] = mark

# Calculating total and percentage
total = sum(marks.values())
percentage = total / len(subjects)

# Calculating CGPA
cgpa = percentage / 9.5

# Assigning Grade
if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "F"

# Pass/Fail Status
if all(mark >= 33 for mark in marks.values()):
    status = "PASS"
else:
    status = "FAIL"

# Display Report
print("\n========== STUDENT REPORT ==========")
print(f"Student Name : {name}")
print(f"Roll Number  : {roll}")

print("\nMarks Obtained:")
for subject, mark in marks.items():
    print(f"{subject}: {mark}")

print("------------------------------------")
print(f"Total Marks : {total}/{len(subjects)*100}")
print(f"Percentage  : {percentage:.2f}%")
print(f"CGPA        : {cgpa:.2f}")
print(f"Grade       : {grade}")
print(f"Result      : {status}")
print("====================================")
```
